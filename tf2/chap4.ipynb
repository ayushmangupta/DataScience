### Law of Large Number

```python
poi = tfp.distributions.Poisson(rate = 4.5)
samples = 100000

for _ in range(3):
    data = poi.sample(sample_shape=samples)
    x = np.arange(1,samples,100)
    avg = [tf.reduce_mean(data[:i]) for i in x]
    plt.plot(x,avg,lw=2)

plt.ylim( 4.35, 4.65) 
plt.title( "Convergence of the average of \n random variables to its \
expected value" )
plt.ylabel( "average of $n$ samples" )
plt.xlabel( "# of samples, $n$")
plt.legend();
```
