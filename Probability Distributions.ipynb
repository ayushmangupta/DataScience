### Poisson distribution

> Another discrete distribution, the Poisson distribution expresses the probability of a given number of events occurring in a fixed interval of time and/or space if these events occur with a known average rate and independently of the time since the last event. Defined for all the integer positive number as,

```python
#colors = matplotlib.rcParams['axes.color_cycle']
plt.figure(figsize=(14,5))

lambda_ = [1,4,5,12]
colorr=cm.rainbow(np.linspace(0,1,len(lambda_)))
k = np.arange(20)
for i,l in enumerate(lambda_):
    #ax = plt.subplot(1,len(lambda_),i+1)
    y = stats.poisson.pmf(k,l)
    #ax.xaxis.set_ticks(k)
    #plt.ylim((0,1))
    #print(y)
    plt.bar(k,y,label=l,color=colorr[i],alpha=0.4,edgecolor=colorr[i], lw=3)
    plt.legend()
    plt.xticks(lambda_)
    #plt.subplots_adjust(bottom=0.15, wspace=2)
plt.show()
```

```python
plt.figure(figsize=(14,5))

x= np.linspace(-10,20,200)
for c,mu,sigma in zip(colorr,[0,1,4,7],[1,3,5,6]):
    y = stats.norm.pdf(x,mu,sigma)
    plt.plot(x,y,color=c,label='$ \mu  =,{} \sigma =  {}$'.format(mu,sigma))
    plt.fill_between(x,y1=0,y2=y,alpha=0.4,color=c)
plt.legend()
plt.show()
```

### Alpha Distribution

```python
x = np.linspace(0.1, 2, 100)
alpha = stats.alpha
alphas = [0.5, 1, 2, 4]
plt.figure(figsize=(12,6))
for a,c in zip(alphas,colorr):
    
    label=r"$\alpha$ = {0:.1f}".format(a)
    plt.plot(x, alpha.pdf(x, a), lw=2, 
              color=c, label=label)
    plt.fill_between(x, alpha.pdf(x, a), color=c, alpha = .33)
    

plt.ylabel("PDF at $x$")
plt.xlabel("$x$")
plt.title("Alpha distribution")

q=plt.legend()
```

### Beta Distribution

```python
beta = stats.beta
x = np.linspace(0,1, num=200)


fig = plt.figure(figsize=(12,6))
for a, b, c in zip([0.5, 0.5, 1, 2, 3], [0.5, 1, 3, 2, 5], colorr):
    plt.plot(x, beta.pdf(x, a, b), lw=2, 
             c=c, label = r"$\alpha = {0:.1f}, \beta={1:.1f}$".format(a, b))
    plt.fill_between(x, beta.pdf(x, a, b), color=c, alpha = .1)
    
    
plt.legend(loc=0)
plt.ylabel("PDF at $x$")
plt.xlabel("$x$")
```

### Exponential Distribution
 the random variable X has an exponential distribution with parameter $\lambda$, we say $X$ is exponential and write $ XExp(\lambda)$ Given a specific $\lambda$, the expected value of an exponential random variable is equal to the inverse of $\lambda$, that is: $E[X|\lambda]=\frac{1}{\lambda}$

```python
def linear_function(x,a,b):
    return a + (x*b)
```
