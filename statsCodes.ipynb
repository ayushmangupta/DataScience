```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import pymc3 as pm
import scipy.stats as stats
import seaborn as sns
plt.style.use('seaborn-darkgrid')
plt.rc('font', size=12)

%config Inline.figure_formats = ['retina']
```

```python
x = np.linspace(0.01,20.0,100)
```

```python
plt.figure(figsize=(12,4))
plt.title("Gamma Distribution")

plt.plot(x,stats.gamma.pdf(x = x,a=1.),label=r"$\theta =  1.0$")
plt.plot(x,stats.gamma.pdf(x = x,a=9.0),label=r"$\theta =  9$")
plt.plot(x,stats.gamma.pdf(x = x,a=2),label=r"$\theta =  2$")
plt.plot(x,stats.gamma.pdf(x = x,a=5.0),label=r"$\theta =  5$")
plt.plot(x,stats.gamma.pdf(x = x,a=3.0),label=r"$\theta =  3$")
#plt.plot(x,stats.gamma.pdf(x = x,a=2))
plt.legend()
```

```python
plt.title("Chi")
#plt.plot(x,stats.chi.pdf(x=x,df=.5),label=r"$df = .5$")
plt.plot(x,stats.chi.pdf(x=x,df=1),label=r"$df = 1$")
plt.plot(x,stats.chi.pdf(x=x,df=2),label=r"$df = 2$")
plt.plot(x,stats.chi.pdf(x=x,df=3),label=r"$df = 3$")
plt.plot(x,stats.chi.pdf(x=x,df=5),label=r"$df = 5$")
plt.plot(x,stats.chi.pdf(x=x,df=10),label=r"$df = 10$")
plt.plot(x,stats.chi.pdf(x=x,df=10),label=r"$df = 15$")

plt.legend()
```
